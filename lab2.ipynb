Метадам `Рунгэ-Кута 3 і 4 парадкаў` знайсці прыблізнае рашэнне задачы Кашы: 
<br>
а) $u'(x)=\frac{1}{2}u^2+x+1, x\in[0;1], u(0)=1$

```python
import numpy as np
import pandas as pd
```

```python
def runge_kutta(f, interval, initial, n=4, step=0.01):
    x = np.arange(interval[0], interval[1] + step, step)
    y = np.zeros(len(x))
    y[0] = initial
    if n == 3:
        for i in range(1, len(x)):
            k1 = step * f(x[i-1], y[i-1])
            k2 = step * f(x[i-1] + step / 2, y[i-1] + k1 / 2)
            k3 = step * f(x[i-1] + step, y[i-1] - k1 + 2* k2)
            y[i] = y[i-1] + (k1 + 4 * k2 + k3) / 6
        return pd.DataFrame({'x': x, 'y': y})
    elif n == 4:
        for i in range(1, len(x)):
            k1 = step * f(x[i-1], y[i-1])
            k2 = step * f(x[i-1] + step / 2, y[i-1] + k1 / 2)
            k3 = step * f(x[i-1] + step / 2, y[i-1] + k2 / 2)
            k4 = step * f(x[i-1] + step, y[i-1] + k3)
            y[i] = y[i-1] + (k1 + 2 * k2 + 2 * k3 + k4) / 6
        return pd.DataFrame({'x': x, 'y': y})
    else:
        print('Not implemented for this dimension')
```

```python
rk3_1 = runge_kutta(lambda x, u: 0.5 * u ** 2 + x + 1, (0, 1), 1, n=3)
rk3_2 = runge_kutta(lambda x, u: 0.5 * u ** 2 + x + 1, (0, 1), 1, n=3, step=0.1)
result1 = pd.DataFrame({'x': rk3_2['x'], 'y, step=0.1': rk3_2['y'], 'y, step=0.01': np.array(rk3_1['y'][::10])})
result1
```

```python
rk4_1 = runge_kutta(lambda x, u: 0.5 * u ** 2 + x + 1, (0, 1), 1, n=4)
rk4_2 = runge_kutta(lambda x, u: 0.5 * u ** 2 + x + 1, (0, 1), 1, n=4, step=0.1)
result2 = pd.DataFrame({'x': rk4_2['x'], 'y, step=0.1': rk4_2['y'], 'y, step=0.01': np.array(rk4_1['y'][::10])})
result2
```
