З дапамогай інтэрпаляцыйнага метаду Адамса ($ q=2 $) знайсці прыблізнае рашэнне задачы Кашы на сетцы вузлоў з $ h=0.1 $, пабудаваўшы пачатак табліцы з карыстаннем метада Рунге-Куты. Для знаходжання $ y_{n+1} $ пабудаваць вылічальны працэс тыпу прэдыктар-карэктар, узяўшы ў якасці прэдыктару адпаведны экстрапаляцыйны метад Адамса. Рэзультаты вылічэнняў прадставіць у выглядзе табліцы значэнняў $ y_{n+1} $, атрыманых па формулах прэдыктар і карэктар, пры наступных уваходных дадзеных:
$$ u'=\frac{e^x}{u}-1, \ u(1)=0.5, \ x\in[1;2,5] $$

```python
import numpy as np
import pandas as pd
```

```python
def f(x, u):
    return np.exp(x) / u - 1
```

```python
def adams(f, interval, initial, step=0.1):
    x = np.arange(interval[0], interval[1] + step / 2, step)
    y = np.zeros(len(x))
    yp = np.zeros(len(x))
    y[0] = initial
    yp[0] = initial
    for i in range(1, 3):
        k1 = step * f(x[i-1], y[i-1])
        k2 = step * f(x[i-1] + step / 2, y[i-1] + k1 / 2)
        k3 = step * f(x[i-1] + step / 2, y[i-1] + k2 / 2)
        k4 = step * f(x[i-1] + step, y[i-1] + k3)
        yp[i] = y[i-1] + (k1 + 2 * k2 + 2 * k3 + k4) / 6
        y[i] = yp[i]
    for i in range(3, len(x)):
        yp[i] = y[i-1] + step / 12 * (23 * f(x[i-1], y[i-1]) - 
                                     16 * f(x[i-2], y[i-2]) + 
                                     5 * f(x[i-3], y[i-3]))
        y[i] = y[i-1] + step / 24 * (9 * f(x[i], yp[i]) + 
                                     19 * f(x[i-1], y[i-1]) - 
                                     5 * f(x[i-2], y[i-2]) + 
                                     f(x[i-3], y[i-3]))
    return x, yp, y
```

```python
x, yp, y = adams(f, (1, 2.5), 0.5)
data = pd.DataFrame({'x': x, 
                     'Predictor, y': yp, 
                     'Corrector, y': y})
```
