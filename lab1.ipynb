Метадам Пікара знайсці прыблізнае рашэнне $y_{n}(x)$ _(n=4)_ задачы Кашы: 
<br>
а) $u'(x)=\frac{1}{2}u^2+x+1, x\in[0;1], u(0)=1$

```python
from sympy import integrate, Symbol, lambdify, sympify, diff, init_printing
import numpy as np
```

```python
def f(x, u):
    return 0.5 * u ** 2 + x + 1
```

```python
x = Symbol('x')
u = [None] * 5
u[0] = diff(x)
for i in range(1, 5):
    u[i] = u[0] + integrate(f(x, u[i-1]))
```

```python
init_printing()
u[4]
```
